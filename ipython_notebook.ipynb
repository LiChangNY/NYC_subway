We can decide whether to recode, transform, or normalize data. For exampele, we need to specify that [DATEn] is read in the correct time format. 

```python
import pandas as pd
import numpy as np
from datetime import datetime

# import data and combined DATEn and TIMEn to datetime
df = pd.read_csv('turnstile_data_master_with_weather.csv')
#df = pd.read_csv('turnstile_data_master_with_weather.csv', parse_dates = [[2,3]]) 
# Recode variables
df.rain = df.rain.astype(int)
df["Raining?"] = np.where(df["rain"] == 1, "Raining", "Not raining")
df['DATEn_TIMEn'] = df['DATEn'] + " " + df['TIMEn']
df['DATEn_TIMEn'] = pd.to_datetime(df.DATEn_TIMEn)
df['Weekday'] = df['DATEn'].apply(lambda d: datetime.strptime(d, '%Y-%m-%d').strftime('%A'))
df['WeekdayN'] = df['DATEn'].apply(lambda d: datetime.strptime(d, '%Y-%m-%d').weekday())
df.head()
```

### 2. T-test

By charting the distribution of ENTRIESn_hourly, we can see that our data do not follow a normal distribution. For this reason, I chose to use non-parametric test: Mann-Whitney U Test.

```python
f = ggplot(df, aes(x='ENTRIESn_hourly', color = 'Raining?')) 
f + geom_density() + \
    ggtitle('Distribution of Entries by Rain') + \
    facet_grid('Raining?')
```

```python
import scipy.stats
Rain = df[(df.rain == 1)]
No_Rain = df[(df.rain == 0)]
u, p = scipy.stats.mannwhitneyu(Rain.ENTRIESn_hourly, No_Rain.ENTRIESn_hourly)
print "p-value: %.2f" %(p * 2)
```

The reported p-value under Mann Whitney test is one-sided, so we need to multiply by two to get the two-sided p-value. The result shows that the difference in total entrie between rainy days and non-rainy days is statistically significant with p-value at 0.05. I can proceed with modelling.

### 3. Modeling

#### Fitting OLS model

```python
###ADD MATH EQUATION
import statsmodels.formula.api as sm
import scipy

def normalize_features(feature):
   feature_normalized = (feature-feature.mean())/feature.std()
   mu = feature.mean()
   sigma = feature.std()
   return feature_normalized

y = df['ENTRIESn_hourly']
df['Eins'] = np.ones((len(df), )) 
X = df[['rain','fog','thunder','meanwindspdi','precipi','Hour','meantempi','meanpressurei', 'Eins']]
#X = normalize_features(X)
dummy_day = pd.get_dummies(df['Weekday'])
X = X.join(dummy_day)
dummy_units = pd.get_dummies(df['UNIT'],prefix='unit')
X = X.join(dummy_units)
```

```python
result = sm.OLS(y, X).fit()
print result.summary()
```

R-square equals 0.566.
It takes IPthon a couple minutes to print out the summary. 

The results suggest that once controlling weekday and turnstile UNIT, 'rain', 'fog', 'precipi', 'Hour', 'meantempi', and 'meanpressure' are signficantly correlated with total entries. So it means that rain weather does play a role in determining the ridership. Interestingly, the coefficient shows that, supposing other factors being the same, e.g. same workday and same UNIT, etc, a raining day will have about 70 less people in entering the subway than a non-raining day. However, if the day turns out to be foggy, there will be about 103 people more entering the subway, which is quite understandable because driving or biking in the foggy day could be dangerous so people are more likely to turn to subway.
